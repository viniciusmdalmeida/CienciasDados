## Autocorrelação

É o estudo de "correlações" entre uma série temporal $Y_t$ em relação à um atraso dessa mesma série $Y_{t-k}$, $k=0,1,2,\dots$. 

### Lag Plot

Se $Y_t$ é uma série temporal, então o lag plot é um método simples de visualizar a influência que $Y_{t}$ tem em $Y_{t+k}$, para $k=0,1,2,\dots$. Para essa análise deve-se fazer um gráfico simples de $Y_t$ por $Y_{t+k}$.

```python
tmp = pbr_cl.AdjClose[1:]
tmp.head()
```

```python
tmp.index = pbr_cl.index[:len(pbr)-1]
```

```python
tmp.head()
```

```python
acr = pd.DataFrame([pbr_cl.iloc[:len(pbr_cl)-1].AdjClose,tmp])
acr.head()
```

```python
acr = acr.transpose()
acr.head()
```

```python
acr.columns = ['Valor','Shift']
```

```python
acr.head()
```

O dataframe acima poderia ser criado mais facilmente com o método `.shift` das séries do pandas:

```python
pbr_cl.AdjClose.shift(1).head()
```

```python
x = np.array(acr.Valor)
y = np.array(acr.Shift)
plt.figure(figsize=(6, 6), dpi=80)
ax = plt.scatter(x=x,y=y)
plt.xlabel(r'$y_t$',fontsize=20)
plt.ylabel(r'$y_{t+1}$',fontsize=20)
```

A Matplotlib já possui uma uma função para fazer esse gráfico automaticamente: `pandas.tools.plotting.lag_plot()`

```python
plt.figure(figsize=(6, 6), dpi=80)
ax = pd.plotting.lag_plot(pbr_cl.loc[pbr_cl.index>=dt.datetime(2018,10,1)].AdjClose,c='red')
ax.set_xlabel(r'$y_t$',fontsize=20)
ax.set_ylabel(r'$y_{t+1}$',fontsize=20)
```

A "correlação" entre as duas colunas pode ser determinada com o método `.corr()` do dataframe:

```python
acr.corr(method='kendall') # possui os seguintes métodos ‘pearson’, ‘kendall’ ou ‘spearman’
```

## Autocorrelação com Statsmodels

```python
from statsmodels.graphics.tsaplots import plot_acf
```

Usaremos os seguintes parâmetros da função `plot_acf`:
 - `x`: o `array` dos dados;
 - `ax`: eixo do matplotlib; 
 - `lags`: `int` ou `array` dos lags a serem usados;
 - `alpha`: intervalo de confiânca (padrão = `0.95`)

```python
Shift_10 = pbr_cl.AdjClose.shift(10)
pbr_cl['Shift_10'] = pbr_cl.AdjClose.shift(10)
```

```python
fig,axes = plt.subplots(nrows=2,ncols=1)
fig.set_figheight(8)
fig.set_figwidth(8)
atraso = np.arange(1,1000,10)
ax_1 = pbr_cl.plot(kind='scatter',x='AdjClose',y='Shift_10',ax=axes[1],c='blue',title='Lag Plot p/ shift=10')
plot_acf(pbr_cl.AdjClose,ax=axes[0],lags=atraso,use_vlines=True,alpha=0.80)
```

## Autoregressão

Usaremos a classe `AR` da `statsmodel` para modelar a autoregressão

```python
pbr2.head()
```

```python
from statsmodels.tsa.ar_model import AR
```

```python
Valores = pbr2.AdjClose.values
Valores
```

```python
len(Valores)
```

```python
Val_treino = Valores[:len(Valores)-3]
len(Val_treino)
```

```python
Val_teste = Valores[len(Valores)-3:]
```

```python
modelo = AR(Val_treino) # cria a modelagem AR(3)
```

```python
modelo.
```

```python
modelo_fit = modelo.fit()
```

```python
previsao = modelo_fit.predict(start=len(Val_treino),end=len(Val_treino)+len(Val_teste)-1)
previsao
```
